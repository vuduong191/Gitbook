```python
import os
import pandas as pd
```

```python
lat=pd.read_csv('lat.csv')['Lat']
long=pd.read_csv('long.csv')['Long']
```

```python
from itertools import product
```

```python
c = list(product(lat, long))
```

```python
data=pd.DataFrame(c,columns=['lat','long'])
```

```python
#cut the lower corner of the rectangular (it's the sea)
data[~((data['lat']<34.001001) & (data['long']<-118.493014))].shape
```
